# Question 1: Explain the difference between Absolute, Relative, and Mixed Cell Referencing in Excel with examples.

## Answer

Cell referencing is one of the most important concepts in Microsoft Excel. It helps us refer to the value of a particular cell while creating formulas. There are three main types of cell references in Excel: Relative, Absolute, and Mixed references.

### 1. Relative Cell Reference

A relative reference changes automatically when a formula is copied from one cell to another.

Example:

If cell A1 contains 10 and B1 contains 20, then:

```excel
=A1+B1
```

When this formula is copied to the next row, it automatically becomes:

```excel
=A2+B2
```

Relative referencing is useful when performing the same calculation for multiple rows or columns.

### 2. Absolute Cell Reference

An absolute reference remains fixed even when the formula is copied to another location. The dollar sign ($) is used to make a reference absolute.

Example:

```excel
=A1*$D$1
```

Here, D1 remains fixed even if the formula is copied to other cells.

Absolute references are commonly used for tax rates, discounts, and constants.

### 3. Mixed Cell Reference

A mixed reference keeps either the row or the column fixed.

Examples:

```excel
=$A1
```

Column A is fixed, but the row can change.

```excel
=A$1
```

Row 1 is fixed, but the column can change.

Mixed references are useful in multiplication tables and matrix calculations.

### Conclusion

Relative references change automatically, absolute references remain fixed, and mixed references partially remain fixed. Understanding these references helps create efficient and error-free Excel formulas.


# Question 2: What is a Macro in Excel? How does it help in automation?

## Answer

A Macro in Excel is a set of recorded actions or instructions that can be executed automatically whenever needed. Macros are created using VBA (Visual Basic for Applications) and are mainly used to automate repetitive tasks.

For example, if a user formats a report every day by applying colors, borders, and formulas, these actions can be recorded as a Macro and executed with a single click.

### Benefits of Macros

1. Saves time by automating repetitive tasks.
2. Reduces manual effort.
3. Minimizes human errors.
4. Increases productivity.
5. Ensures consistency in reports.

### Example

Suppose a company receives sales data daily. Instead of formatting the report manually every day, a Macro can automatically:

* Apply formatting
* Create charts
* Generate summaries
* Highlight important values

### Steps to Create a Macro

1. Go to the Developer tab.
2. Click Record Macro.
3. Perform the desired actions.
4. Click Stop Recording.
5. Run the Macro whenever required.

### Conclusion

Macros are powerful Excel tools that automate repetitive tasks and improve efficiency. They help users save time and increase accuracy while working with large datasets.


# Question 3: What are Text Functions in Excel? Mention any five with examples.

## Answer

Text functions in Excel are used to manipulate, combine, extract, or modify text data. They are very useful when working with names, addresses, phone numbers, and other textual information.

### 1. LEFT Function

Returns a specified number of characters from the left side of a text string.

Example:

```excel
=LEFT("Excel",2)
```

Result: Ex

### 2. RIGHT Function

Returns characters from the right side of a text string.

Example:

```excel
=RIGHT("Excel",3)
```

Result: cel

### 3. MID Function

Returns characters from the middle of a text string.

Example:

```excel
=MID("Excel",2,3)
```

Result: xce

### 4. LEN Function

Returns the total number of characters in a text string.

Example:

```excel
=LEN("Excel")
```

Result: 5

### 5. CONCAT Function

Combines multiple text strings into one.

Example:

```excel
=CONCAT("Data"," ","Analyst")
```

Result: Data Analyst

### Importance of Text Functions

* Data cleaning
* Data formatting
* Report preparation
* Customer data management

### Conclusion

Text functions help users efficiently manage and manipulate text data in Excel. They are widely used in data analysis and reporting tasks.

Question 5: Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?
Answer

Microsoft Excel provides several lookup functions that help users search for specific information from large datasets. Among these, VLOOKUP, HLOOKUP, XLOOKUP, and INDEX-MATCH are the most commonly used functions. These functions save time and make data retrieval faster and more accurate.

VLOOKUP

VLOOKUP stands for Vertical Lookup. It is used when data is arranged vertically in columns. It searches for a value in the first column of a table and returns a corresponding value from another column in the same row.

Syntax:

=VLOOKUP(lookup_value, table_array, col_index_num, FALSE)

Example:

=VLOOKUP(101,A2:D10,3,FALSE)

In this example, Excel searches for Employee ID 101 in the first column and returns the value from the third column.

HLOOKUP

HLOOKUP stands for Horizontal Lookup. It is similar to VLOOKUP, but it works with data arranged horizontally in rows. It searches for a value in the first row and returns a corresponding value from another row.

Syntax:

=HLOOKUP(lookup_value, table_array, row_index_num, FALSE)

Example:

=HLOOKUP("Sales",A1:H5,3,FALSE)

This formula searches for the value "Sales" in the first row and returns the value from the third row.

XLOOKUP

XLOOKUP is a modern and advanced lookup function introduced in newer versions of Excel. It overcomes many limitations of VLOOKUP and HLOOKUP. XLOOKUP can search both vertically and horizontally and can return values from any direction.

Syntax:

=XLOOKUP(lookup_value, lookup_array, return_array)

Example:

=XLOOKUP(101,A2:A10,C2:C10)

This formula searches for Employee ID 101 in column A and returns the corresponding value from column C.

````python
## QUESTION 6
### Importing required libraries
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
````

````python
# Step 1: Creating a dummy dataset of 8 employees with their joining dates

data = {
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107, 108],
    "EmployeeName": ["Aarav Sharma", "Priya Singh", "Rohan Mehta", "Sneha Patel",
                     "Vikram Rao", "Ananya Das", "Karan Gupta", "Isha Verma"],
    "JoiningDate": ["2018-03-15", "2020-07-01", "2015-11-23", "2022-01-10",
                     "2019-05-30", "2021-09-05", "2023-06-18", "2016-12-01"]
}

# Converting dictionary to a DataFrame
df = pd.DataFrame(data)

# Converting JoiningDate column from text to actual date format
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"])

# Displaying the dataset
df
````

````python
# Step 2: Calculating experience using today's date

today = datetime.today()

# Creating empty lists to store years and months
exp_years = []
exp_months = []

# Looping through each employee's joining date
for date in df["JoiningDate"]:
    diff = relativedelta(today, date)   # gives difference in years, months, days
    exp_years.append(diff.years)
    exp_months.append(diff.months)

# Adding results as new columns in the dataframe
df["Experience_Years"] = exp_years
df["Experience_Months"] = exp_months

# Combining into one readable column
df["Total_Experience"] = df["Experience_Years"].astype(str) + " years " + df["Experience_Months"].astype(str) + " months"

# Final output
df
````

````python
print("Employee Experience Report (as on", today.date(), ")\n")
print(df[["EmployeeID", "EmployeeName", "JoiningDate", "Total_Experience"]])
````
